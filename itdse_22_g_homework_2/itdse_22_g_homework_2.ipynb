### Q0
**Fix the format** of the *Neighborhood"* as described in cleaning instructions. **If there will be other automatic cleaning include in below as a new cell.** Below, you will see one more question, **explain your both manual and automatic (if there is any) in that cell _briefly_.**

```python
df['Neighborhood'] = df['Neighborhood'].str.capitalize().str.translate(str.maketrans("ğĞıİöÖüÜşŞçÇ", "gGiIoOuUsScC")).str.extract(pat=r'([A-Z][a-z]+)')
df['Neighborhood']
```

# First **capitalize** all neighborhoods and then **replace** turkish characters with english characters finally **select** only the first word from neighborhood description.
# For example ayazağa(kampüs) entry for Neighborhood column was invalid and I handled the data by doing this ^^
# **capitalize** ==> Ayazağa(kampüs)
# **replace** ==> Ayazaga(kampus)
# **select** ==> Ayazaga

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)   # Set NaN values to column's mode
df['GPA'].fillna(value=df['GPA'].mean(), inplace=True)   # Set NaN values to column's mean
df['Repeat_Course_Num'].fillna(df['Repeat_Course_Num'].mode()[0], inplace=True)   # Set NaN values to column's mode
df['Expected_GPA'].fillna(value=df['Expected_GPA'].mean(), inplace=True)   # Set NaN values to column's mean
df['Expected_Letter_Grade'] = df['Expected_Letter_Grade'].str.upper()   # I treat the invalid string values. For example bb entry for Expected_Letter_Grade column was invalid and I handled the data by doing str.upper()
df['Expected_Letter_Grade'].fillna(df['Expected_Letter_Grade'].mode()[0], inplace=True)   # Set NaN values to column's mode
df['Neighborhood'].fillna(df['Neighborhood'].mode()[0], inplace=True)   # Set NaN values to column's mode
df['Latitude'].fillna(value=df['Latitude'].mean(), inplace=True)   # Set NaN values to column's mean
df['Longitude'].fillna(value=df['Longitude'].mean(), inplace=True)   # Set NaN values to column's mean
df['Travel_Time'].fillna(value=df['Travel_Time'].mean(), inplace=True)   # Set NaN values to column's mean
df['Time_For_Study'].fillna(value=df['Time_For_Study'].mean(), inplace=True)   # Set NaN values to column's mean
df['Time_For_Hobbies'].fillna(value=df['Time_For_Hobbies'].mean(), inplace=True)   # Set NaN values to column's mean
df['Time_For_Sports'].fillna(value=df['Time_For_Sports'].mean(), inplace=True)   # Set NaN values to column's mean
df['Time_For_Socializing'].fillna(value=df['Time_For_Socializing'].mean(), inplace=True)   # Set NaN values to column's mean
df['Time_For_Entertainment'].fillna(value=df['Time_For_Entertainment'].mean(), inplace=True)   # Set NaN values to column's mean

df
```

#### Part 1
Show the frequency distribution of the GPAs by a plot. **Do not show** outlier/s (**extremely low values**) for better visualization. **Show values only between 1-4 for this plot**. Explain what you have understand from the graph in few sentences.

```python
plt.figure(figsize=(12,8))
plt.title("Frequency distribution of the GPAs",fontsize=20)
plt.xlim(1,4)
plt.hist(df['GPA'], density=True, histtype='stepfilled', alpha=0.3, bins=50,label="GPA", color='green')
sns.kdeplot(df['GPA'], color='green')
plt.xlabel("GPA",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

# Almost everyone's GPA's are above 2.0.
# There is intensity in the interval from 2.5 to 3.5.
